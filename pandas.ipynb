```python
#Merging 2 data frames
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})

right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})

mergedDf = pd.merge(left, right, on="key")
mergedDf
```

```python
#grouping 
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)

df.groupby("B")[["C", "D"]].sum() # it specifies after grouping want tonly C, D columns as they are like integers and sum operation can be made.
```

```python
#handling categorical values
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)
df["grade"] = df["raw_grade"].astype("category")

df["grade"]
```

```python
#we can replace categorical values with some meaningful names
categoricalReplacements = ["Good", "OK", "Bad"]
df["grade"].cat.rename_categories(categoricalReplacements)
```
