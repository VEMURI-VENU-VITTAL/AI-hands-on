```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel("Values")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 5, 7, 9])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
t = np.arange(0,5,0.2)
plt.plot(t, t, "r--", t, t**2, "b^", t, t**3, "go")
plt.show()
```

```python
#scatter plt

data = {
    "a":np.arange(0,50),
    "c":np.random.randint(0,50, 50),
    "d":np.random.randn(50)
}

data["b"] = data["a"]+10*np.random.randn(50)
data["d"] = data["d"]*100

plt.scatter("a", "b", c="c", s="d", data=data)
plt.show()

'''
c -> color map as per value color will be decided to point
d -> it means distribution, decides how big the point can be
Scatter plot have an option that takes dictionary called data from that it searches arguments that are required
'''
```

```python
#plotting multiple graphs in a single frame
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()

'''
subplot accepts 3 digit number as value, first digit represents number of rows, second digit represents number of columns, third digit represent order of the sub plot 
'''
```

```python
#setp method will allow to set properties to the graph
x = [2, 20, 100]
y = [10, 200, 1000]
lines = plt.plot(x, y)
properties = plt.setp(lines, color='r', linewidth='2.0')
plt.show()
```
