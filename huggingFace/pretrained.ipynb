```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "distilgpt2"
tokernizer = AutoTokenizer.from_pretrained(
    model_name
)

model = AutoModelForCausalLM.from_pretrained(
    model_name
)
```

```python
text = "Artificial Intelligence is"

tokens = tokernizer(
    text,
    return_tensors="pt"
)

print(tokens)
```

```python
output = model.generate(
    **tokens,
    max_new_tokens=50
)
```

```python
output
```

```python
print(tokernizer.decode(
    output[0],
    skip_special_tokens=True
))
```

```python
from sentence_transformers import SentenceTransformer
from torch.nn.functional import cosine_similarity
import torch

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

s1 = model.encode(
    "I love machine learning",
    device="cpu"
)

s2 = model.encode(
    "I enjoy artificial intelligence",
    device="cpu"
)

s3 = model.encode(
    "The weather is sunny",
    device="cpu"
)

s1 = torch.tensor(s1)
s2 = torch.tensor(s2)
s3 = torch.tensor(s3)

print(
    cosine_similarity(
        s1.unsqueeze(0),
        s2.unsqueeze(0)
    )
)

print(
    cosine_similarity(
        s1.unsqueeze(0),
        s3.unsqueeze(0)
    )
)
```
